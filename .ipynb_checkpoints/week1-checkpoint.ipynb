```python
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Year'] = pd.DatetimeIndex(df['Date']).year
```

```python
df['PxV'] = df['Close Price'] * df['Total Traded Quantity']
```

```python
vwap = []
for month,df2 in month_group:
    vwap.append(sum(df2['PxV'])/sum(df2['Total Traded Quantity']))
vwap
```

```python
def average_price(n):
    return df['Close Price'][-n:].mean()

def profit_loss_percent(n):
    return ((df['Close Price'].iloc[-1] - df['Close Price'].iloc[-n])/df['Close Price'].iloc[-1])*100
```

```python
print("Average Price over Last:\n\n1 week : {}\n2 weeks: {}\n1 month: {}\n3 months: {}\n6 months: {}\n1 year: {}".format(average_price(7),average_price(14),average_price(30),average_price(90),average_price(180),average_price(365)))

print("\n\nProfit/Loss % over Last:\n\n1 week : {}\n2 weeks: {}\n1 month: {}\n3 months: {}\n6 months: {}\n1 year: {}".format(profit_loss_percent(7),profit_loss_percent(14),profit_loss_percent(30),profit_loss_percent(90),profit_loss_percent(180),profit_loss_percent(365)))
```

```python
df['Day_Perc_Change'] = df['Close Price'].pct_change()
```

```python
df['Day_Perc_Change'][0] = 0
df.head()
```

```python
def trend(n):
    if n>-0.5 and n<0.5:
        return "Slight or No change"
    elif n>0.5 and n<1:
        return "Slight positive"
    elif n>-1 and n<-0.5:
        return "Slight negative"
    elif n>1 and n<3:
        return "Positive"
    elif n>-3 and n<-1:
        return "Negative"
    elif n>3 and n<7:
        return "Among top gainers"
    elif n>-7 and n<-3:
        return "Among top loosers"
    elif n>7:
        return "Bull run"
    elif n<-7:
        return "Bear drop"
    
df['Trend'] = df['Day_Perc_Change'].apply(trend)
```

```python
trend_group = df.groupby('Trend')
print("Average")
for i,j in trend_group:
    print(i,end = " :")
    print(j['Total Traded Quantity'].mean())
print("Median")
for i,j in trend_group:
    print(i,end = " :")
    print(j['Total Traded Quantity'].median())
```
